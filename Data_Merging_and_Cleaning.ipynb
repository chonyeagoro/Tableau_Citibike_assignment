```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px 
import numpy as np
import requests
import json
import csv
```

# Reading in and Merging CSV files 

```python
citibike_jan_df = pd.read_csv("202301-citibike-tripdata_1.csv")
citibike_feb_df = pd.read_csv("202302-citibike-tripdata_1.csv")
citibike_mar_df = pd.read_csv("202303-citibike-tripdata_1.csv")
citibike_apr_df = pd.read_csv("202304-citibike-tripdata_1.csv")
citibike_may_df = pd.read_csv("202305-citibike-tripdata_1.csv")
citibike_jun_df = pd.read_csv("202306-citibike-tripdata_1.csv")
citibike_jul_df = pd.read_csv("202307-citibike-tripdata_1.csv")
citibike_aug_df = pd.read_csv("202308-citibike-tripdata_1.csv")
citibike_sept_df = pd.read_csv("202309-citibike-tripdata_1.csv")
citibike_oct_df = pd.read_csv("202310-citibike-tripdata_1.csv")
citibike_nov_df = pd.read_csv("202311-citibike-tripdata_1.csv")
citibike_dec_df = pd.read_csv("202312-citibike-tripdata_1.csv")

citibike_jan_df
```

```python
citibike_2023_data_merged_df = pd.concat([citibike_jan_df, citibike_feb_df, citibike_mar_df,
                                          citibike_apr_df, citibike_may_df, citibike_jun_df,
                                          citibike_jul_df, citibike_aug_df, citibike_sept_df,
                                          citibike_oct_df, citibike_nov_df, citibike_dec_df],
                                         ignore_index=True)
citibike_2023_data_merged_df
```

```python
citibike_2023_data_merged_df.isnull().any()
```

```python
citibike_2023_data_merged_df.dropna(inplace=True)
citibike_2023_data_merged_df.isnull().any()
```
